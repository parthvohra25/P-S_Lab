question 1

```r
a<-punif(45,0,60,lower.tail=FALSE)    #(a)
print(a)
b<-punif(30,0,60)-punif(20,0,60)      #(b)
print(b)
```

punif is used to find cumulative frequencies. we simply put the value above which we want to find out the probabilities along with min and max values.
if we do not write lower.tail=FALSE then we get the values that are subtracted from 1.

question 2

```r
x<-3
a1<-dexp(x,1/2)       #(a)
print(a1)
x1<-0:5
plot(x1,dexp(x1,1/2))  #(b)
a<-pexp(x,1/2)         #(c)
print(a)
vec<-0:5
plot(vec,pexp(vec,1/2)) #(d)
x_sim<-rexp(1000,1/2)
plot(density(x_sim))    #(e)
```

dexp returns the corresponding values of exponential densities for a given x whereas pexp returns the corresponding values of exponential cummulative distribution function.
rexp is used to stimulate a group of random variables chosen from the exponential distribution

question 3

```r
x<-3                              #(a)
a<-dgamma(x,shape=2,scale=1/3)
print(a)
x1<-1
a1<-pgamma(x1,shape=2,scale=1/3)
print(a1)
p<-0.70
a2<-qgamma(p,shape=2,scale=1/3)
print(a2)
```

dgamma is used to find the probability at a given x.
pgamma is used to find cumulative probability distribution of a given x for gamma distribution.
qgamma finds the value of inverse cumulative distribution function.
