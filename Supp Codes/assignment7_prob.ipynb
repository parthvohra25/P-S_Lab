question 1

```r
n=100
c1=n-1
val<-rt(n,c1)
print(val)
hist(val,,main="HISTOGRAM PLOT",xlab="X",ylab="Y",border="blue",col="pink")
```

rt function is used to create random sequence of values from t distribution and hence plotted the graph for the numbers generated.
t distribution is used to estimate population parameters for small sample sizes or unknown variances.

question 2

```r
n=100
df<-c(2,10,25)
for(i in df)
{
  print(rchisq(n,i))
}
```

rchisq is used to generate a list of n random values that follow a chi square distribution with given degree of freedom.
n=no of values to be generated
i=degree of freedom(df)
chi square distribution is distribution with k degrees of freedom.it is used to describe sum of squared random variables.

question 3

```r
x<-seq(-6,6,length=100)
df<-c(1,4,10,30)
color<-c("orange","green","blue","pink")
for(i in df)
{
  print(dt(x,i))
}
plot(x,dt(x,df[4]),type="l",xlab="t-value",ylab="density",main="comparison of t distribution",col=color[4])
for(i in 1:4)
{
  lines(x,dt(x,df[i]),type="l",col=color[i])
}
```

dt function is used to return probability density analysis of t distribution with random variable x and degree of freedom df.
